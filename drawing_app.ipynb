```python
import ipywidgets as widgets
from IPython.display import display
from PIL import Image, ImageDraw
import matplotlib.pyplot as plt

def buat_gambar(warna="#FF69B4", ukuran=10):
    img = Image.new("RGB", (500, 300), "white")
    draw = ImageDraw.Draw(img)
    draw.line((100, 150, 400, 150), fill=warna, width=ukuran)
    draw.text((10, 280), "PinkBrush di Jupyter ðŸ’š", fill="black")

    plt.figure(figsize=(8, 4))
    plt.imshow(img)
    plt.axis('off')
    plt.show()

# Buat widget
warna_widget = widgets.ColorPicker(
    value='#FF69B4',
    description='Warna:',
    disabled=False
)

ukuran_widget = widgets.IntSlider(
    value=10,
    min=1,
    max=30,
    description='Ukuran:',
    disabled=False
)

ui = widgets.VBox([warna_widget, ukuran_widget])
out = widgets.interactive_output(buat_gambar, {
    'warna': warna_widget,
    'ukuran': ukuran_widget
})

display(ui, out)
```
